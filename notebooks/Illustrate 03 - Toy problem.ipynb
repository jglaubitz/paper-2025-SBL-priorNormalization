# Illustrate Toy Problem 

```julia
# Import module. 
using Revise
using PriorNormalization
```

```julia
# Import packages. 
using CairoMakie # for plots
using LaTeXStrings
#using Plots: Plots, plot, plot!, scatter, scatter!, savefig, surface, surface!
#Plots.pyplot()
```

## Set-up the toy problem

```julia
# Parameter of of generalized gamma hyper-prior 
r_range = [ 1.0, .5, -.5, -1.0 ]; 
β_range = [ 1.501, 3.0918, 2.0165, 1.0017 ];
ϑ_range = [ 5*10^(-2), 5.9323*10^(-3), 1.2583*10^(-3), 1.2308*10^(-4) ]; 
```

```julia
# Fix model parameters.
y = .2 # observational data 
σ² = 10^(-2.8) # noise variance 
```

```julia
# Define the posterior density 
function posterior_pdf(x, θ; y, σ², r, β, ϑ)
    if θ <= 0 
        pdf = 0 
    else
        pdf = θ^(r*β-1.5) * exp( -0.5*(x-y)^2/σ² - 0.5*x^2/θ - (θ/ϑ)^r )
    end
    
    return pdf
end
```

## Contour plot of posterior: $r = 1$

```julia
# Select hyper-hyper-parameters 
i = 1 
r = r_range[i] # power parameter 
β = β_range[i] # shape parameter 
ϑ = ϑ_range[i] # scale parameter 

# Generate a grid of values for x and y
#x_range = range( .1, stop=.275, length=1000 ) 
#θ_range = range( 10^(-5), stop=.155, length=1000 )
x_range = range( -.01, stop=0.275, length=1000 ) 
θ_range = range( 10^(-5), stop=.1, length=1000 )

# Evaluate the posterior PDF at each grid point
pdf_values = [posterior_pdf(x, θ; y, σ², r, β, ϑ) for x in x_range, θ in θ_range]

# Contour plot 
fig = Figure()
ax = Axis(fig[1, 1], xlabel = L"x", ylabel = L"\theta")
ax.xlabelsize=30
ax.ylabelsize=30
ax.xticklabelsize=20
ax.yticklabelsize=20

CairoMakie.contour!(ax, x_range, θ_range, pdf_values; levels=6, linewidth=5) 

display(fig)

# Save the plot
save("contour_posterior_toy_rp1.pdf", fig)  # Save as PDF
```
